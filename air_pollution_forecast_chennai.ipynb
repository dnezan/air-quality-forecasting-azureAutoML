```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df_2015=pd.read_csv('./data/datafile_2015.csv')

df_2015['Sampling Date'] = pd.to_datetime(df_2015['Sampling Date'], dayfirst='True')

#df_2015 = df_2015[(df_2015['Location of Monitoring Station'] == 'Kathivakkam, Municipal Kalyana Mandapam, Chennai') | (df_2015['Location of Monitoring Station'] == 'Govt. High School, Manali, Chennai.') | (df_2015['Location of Monitoring Station'] == 'Thiruvottiyur, Chennai')]
df_2015 = df_2015[(df_2015['Location of Monitoring Station'] == 'Kathivakkam, Municipal Kalyana Mandapam, Chennai')]
df_2015.to_csv('./temp/datefile_2015CLEAN.csv')

# Plot the lines for each location
pivot_df = pd.pivot_table(df_2015, values='SO2', index='Sampling Date', columns='Location of Monitoring Station')
pivot_df = pivot_df.interpolate(method='linear')
pivot_df.plot()
plt.legend()
plt.show()
```
